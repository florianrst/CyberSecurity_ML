```python


# Group and reshape data
hour_attack = (
    df.groupby('hour')['Attack Type']
    .value_counts()
    .unstack(fill_value=0)
)

# Plot stacked bar chart
hour_attack.plot(
    kind='bar',
    figsize=(12, 6),
    stacked=True
)

plt.xlabel('Hour of Day')
plt.ylabel('Number of Attacks')
plt.title('Attack Types Distribution by Hour')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

 At each hour, how many DDoS, Malware, and Intrusion attacks occurred?

```python
hourly_attacks = (
    df.groupby('hour')['Attack Type']
      .value_counts()
      .unstack(fill_value=0)
)
```

```python
hourly_attacks
```

```python
#no. of attacks per hr

plt.figure(figsize=(10,6))

for attack in hourly_attacks.columns:
    plt.plot(hourly_attacks.index, hourly_attacks[attack], label=attack)

plt.xlabel('Hour of Day')
plt.ylabel('Number of Attacks')
plt.title('Hourly Distribution of Attack Types')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#When should defenses be strongest?
hourly_attacks['Total'] = hourly_attacks.sum(axis=1)
hourly_attacks.sort_values('Total', ascending=False).head()
```

```python


# Attack Type vs Action Taken
attack_action = (
    df.groupby(['Action Taken', 'Attack Type'])
      .size()
      .unstack(fill_value=0)
)

attack_action.plot(kind='bar', figsize=(10,6))
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.title('Attack Type vs Action Taken')
plt.legend(title='Action Taken')
plt.tight_layout()
plt.show()
```
